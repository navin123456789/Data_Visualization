```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4]
y=[98,87,92,77]
plt.plot(x,y,color='red',linestyle='dashed',linewidth=2,alpha=0.5,marker='+',markersize=10)
```

```python
plt.plot(x,y,'g*:',linewidth=2)
```

```python
Rollno=[1,2,3,4]
DS=[97,94,86,79]
OOP=[98,96,95,95]
AI=[93,91,85,83]
```

```python

plt.plot(Rollno,DS,'g*--',label='Marks in Data Structure')
plt.plot(Rollno,OOP,'b+:',label='Marks in Object Oriented Programming')
plt.plot(Rollno,AI,'yD-.',label='Marks in Artificial intelligence')
plt.legend(loc='best',shadow=True,fontsize='small')
plt.grid()
##loction of legend can be changed as loc=upper right,best etc
```

```python
Students=['Ayush','Avinav','Navin','Shlok']
GPA1=[3.86,3.72,3.78,3.86]
GPA2=[3.95,3.89,3.86,3.82]
plt.title("Marksheet result")
plt.bar(Students,GPA2,label="GPA in 2nd sem")
plt.bar(Students,GPA1,label="GPA in 1st sem")
plt.legend(loc='lower right')
```

```python
Students=['Ayush','Avinav','Navin','Shlok']
GPA1=[3.86,3.72,3.78,3.86]
GPA2=[3.95,3.89,3.86,3.82]

```

```python
xpos=np.arange(len(Students))
xpos
```

```python
plt.bar(xpos-0.2,GPA1,width=0.4,label="GPA in 1st sem ")
plt.bar(xpos+0.2,GPA2,width=0.4,label="GPA in 2nd sem ")
plt.xticks(xpos,Students)
plt.legend(loc='lower left')
```

```python
blood_sugar_men=[85,120,110,93,110,102,125,82,140,122,100,113,82]
blood_sugar_women=[80,99,105,125,160,82,123,160,97,88]
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,120,150],rwidth=0.95,color=['green','orange'],label=['men','women'])##defalult it has 10 bins
plt.legend(loc='best')
```
